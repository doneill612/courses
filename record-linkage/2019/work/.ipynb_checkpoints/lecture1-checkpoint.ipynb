```python
import unicodedata

def strip_accents(text):

    try:
        text = unicode(text, 'utf-8')
    except NameError: # unicode is a default on python 3 
        pass

    text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore').decode("utf-8")

    return str(text)
```

```python
def cleanstr(val):
    val = val.lower()
    if '(' in val:
        val = val[:val.index('(')]
    return strip_accents(val.strip()).replace(' ', '')

df.artist = df.artist.apply(cleanstr)
```

```python
from collections import defaultdict

def cluster(ser):
    groups = defaultdict(set)
    
    for val in ser.values:
        groups[cleanstr(val)].add(val)
        
    for c in groups.items():
        yield c
```

```python
import string

def tokenize(chars):
    chars = chars.lower().strip()
    chars = chars.translate(str.maketrans('', '', string.punctuation))
    splits = sorted(chars.split(' '))
    return ' '.join([split for split in splits])
    
print(tokenize('University of North Carolina, NC'))
print(tokenize('University of North Carolina NC'))
print(tokenize('University of north Carolina, NC'))
```
